```python
import pandas as pd
import numpy as np
import os
```

```python
df = pd.read_csv('india_housing_prices.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
#Standardize column names
df.columns = [c.strip().replace(' ', '_').replace('-', '_').lower() for c in df.columns]

#Drop exact duplicates
df = df.drop_duplicates()
```

```python
df.dtypes
```

```python
#Fill missing numeric values
num_cols = df.select_dtypes(include=['float64','int64']).columns.tolist()
for c in num_cols:
    df[c] = df[c].fillna(df[c].median())

#Fill categorical missing
cat_cols = df.select_dtypes(include=['object']).columns.tolist()
for c in cat_cols:
    df[c] = df[c].fillna('Unknown')
```

```python
def outlier_treatment(df, col='price_per_sqft'):
    # simple IQR-based cap
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr
    df[col] = df[col].clip(lower=lower, upper=upper)
    return df
df['price_per_sqft'] = (df['price_in_lakhs'] * 100000) / df['size_in_sqft']
df = outlier_treatment(df)
```

```python
def save_clean(df, path):
    os.makedirs(os.path.dirname(path), exist_ok=True)
    df.to_csv(path, index=False)
    print(f"Saved cleaned data to {path}")
save_clean(df, "C:/Users/Dell/Desktop/DATA ANALYTICS/DA_PROJECTS/LA3/cleaned_india_housing_prices.csv")
```
